```python
def Vfunc(ac, t):
#     return 1
    return ac + t
def ac(t):
    m= 2
    return 1.0+math.sin(t*m)+t
    
def X(y,r):
#     print(r*r - y*y)
    return math.sqrt((r*r) - (y*y))
def Y(s, t):
    a = 0.5
    alpha = 2.0
    k = 1.0
    V = Vfunc(ac(t), t)
#     V = 1
    return a*math.exp(alpha*(s-1.0))*math.sin(k*(s-V*t))
# for i in range(10):
#     print(i/10, Y(i/10, 100))
```
