```python
def Y(s, t):
    a = 0.1
    alpha = 2.0
    k = 10.0
    V = 5
    return a*math.exp(alpha*(s-1.0))*math.sin(k*(s-V*t))
def X(s,y):
    assert((s*s - y*y) >= 0)
    return math.sqrt(s*s - y*y)
```
